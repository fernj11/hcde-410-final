## **Overview and Research Questions** ##

COVID-19 has impacted the lives of many and has been ever-present in society since late 2019. Since it became a pandemic there have been several organizations collecting data regarding many different variables connected to the virus and its spread. Because of this, there is a lot of data available that allows for different research questions to be asked and questioned. In this research exploration of the data will be done to answer some questions which are written below. 

It connects well to a human-centered perspective because as mentioned before it has impacted many people across the world. All people can make a connection to the virus and recall a time they, a relative, or a friend had tested positive for it. Making it beneficial to research.

**Questions**

- How do the regulations and quarantining styles of different areas (cities, states, countries) at the beginning of COVID-19 show in the data?
- Which areas had the best initial response to the COVID-19 pandemic?
- What patterns can be seen with the data over the years and change of regulations?
- Are there changes with the variants of the COVID-19?
- How do personal accounts differ from state to state, country to country?

From these questions I would like to focus on the first, which relates to the regulations and quarantining styles of different areas at the beginning of COVID-19 and how it's shown in collected data. From this research I hope to find  patterns within the data and be able to connect it to the events that happened and how they impacted the total cases, vaccinations, etc. 

**Hypothesis**
I believe that areas with stricter regulations had better responses to COVID-19, but there are certain locations where this backfired resulting in worse results.
