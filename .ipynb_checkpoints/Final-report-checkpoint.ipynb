## **Findings** ##

**1. What trends can be observed among Washington counties that contribute to higher case numbers and fatality rates due to COVID-19?**

The overall premise of this study originated from the question of identifying variables that contributed to case numbers and deaths due to COVID-19. However, it is important to note that there are many other variables beyond those selected for this study. The variables chosen were those most closely tied to the effects of COVID-19, particularly as it spread quickly in large populations and disproportionately impacted older age groups. Socioeconomic variables were also included, as they can limit access to healthcare and overall health and wellness.

Using a correlation matrix in Python, the relationships between various variables with fatality rates, and case numbers were calculated. The results revealed varying positive correlations between fatality rates and case numbers. Case numbers appeared most influenced by household size, where counties with larger average household sizes had a higher likelihood of higher case numbers. This was the strongest correlation in the study, with a score of 0.75 on a scale from -1 to 1 (with 1 being a positive correlation and -1 being a negative correlation). Conversely, counties with a larger population of people aged 65 and older had a negative correlation with case numbers, scoring -0.68, indicating that older populations tended to have fewer cases.

Interestingly, some variables showed little correlation with case numbers. These included population density with a score of -0.028 and vaccine coverage with a score of -0.067. These scores suggest that these variables had negligible impact on case numbers. This was surprising, as COVID-19 seemed more influenced by household size than by proximity or larger populations as a whole. It was also notable that while vaccine coverage had little effect on the spread of COVID-19 in counties, it did have a significant impact on fatality rates.

When considering variables that impacted fatality rates, it was no surprise that vaccination coverage was strongly correlated with lower fatality rates, with the highest correlation score of -0.6. Similarly, the population of individuals aged 65 and older had the highest positive correlation with fatality rates, with a score of 0.45, as COVID-19 disproportionately affected older individuals. Following closely was the poverty level, with a score of 0.37, suggesting that areas with higher poverty levels tended to have higher death rates. Population density also emerged as an interesting variable, as higher population density was associated with lower fatality rates, indicating that areas with more urbanization had lower death rates.

**Graph & Code Explanation**

For the use of graphs, the data was loaded as a CSV and named data. The 'Statewide' portion of the data was taken out as it was the total amounts. A new column was also created with the case number percentages. Columns in the dataset were then selected to be analyzed along with the case percentages. After the data was analyzed with the .corr command in Python. From there seaborn and pyplot was used to create a heatmap and the graph titles. The graph displays the correlation (R-value) between the variable (y-axis) with case number percentage correlation (x-axis).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = 'covid_data.csv'  
data = pd.read_csv(file_path)

filtered_data = data[data['county'] != 'Statewide'].copy() 

filtered_data.loc[:, 'cases_percentage'] = (filtered_data['cases_reported'] / filtered_data['population']) * 100

numerical_columns = ['cases_percentage', 'fatality_rate_pct', 'population', 'vaccine_coverage_pct', 
    'population_65_plus_pct', 'poverty_pct', 'population_density', 'household_size', 'hospitals']

correlation = filtered_data[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation[['cases_percentage']].sort_values('cases_percentage', ascending=False),
            annot=True, cmap='coolwarm', cbar=True, vmin=-1, vmax=1)

plt.title("Correlation of Case Percentage with County Variables")
plt.ylabel("Variables")
plt.xlabel("Case Percentage Correlation")
plt.tight_layout()
plt.show()
```

**2. What variables are associated with more cases per population and higher fatality rates?**

**Variables Associated with More Cases per Population**

- **Household Size:** Counties with larger average household sizes had a strong positive correlation (0.75) with higher case numbers.
- **Older Populations:** Counties with a larger population aged 65 and older showed a negative correlation (-0.68), suggesting fewer cases in these areas.
- **Negligible Factors:** Population density (-0.028) and vaccine coverage (-0.067) showed minimal correlation, indicating limited influence on case numbers.

**Variables Associated with Higher Fatality Rates**

- **Vaccine Coverage:** Lower vaccine coverage correlated strongly with higher fatality rates (-0.6).
- **Older Populations:** Areas with a larger population aged 65 and older had a positive correlation (0.45) with fatality rates, as older individuals were more vulnerable to severe outcomes.
- **Poverty Levels:** Higher poverty levels showed a moderate positive correlation (0.37) with increased death rates.
- **Population Density:** Interestingly, higher population density correlated with lower fatality rates, suggesting urban areas experienced fewer fatalities per population.

**3. Does medical availability, such as the presence of hospitals, influence case numbers and fatality rates?**

**Impact on Case Numbers**

A slightly positive trend was observed, with an R-value of 0.23, indicating a weak correlation between the number of hospitals and case percentages. This suggests that having more hospitals in a county does not strongly correlate with the spread of cases.

**Impact on Fatality Rates**

The correlation between hospitals and fatality rates was negative, meaning counties with more hospitals tended to have lower fatality rates. This aligns with the broader finding that access to medical care, such as hospitals, likely reduces the severity and outcomes of COVID-19 infections. However, the relationship was not particularly strong, suggesting other factors, such as vaccine coverage, poverty levels, and demographics, play more significant roles.

**Graph & Code Explanation**

For the scatter plot analysis, data was loaded from a CSV file and named data. The 'Statewide' portion of the data was taken out as it was the total amounts. A new column was also created with the case number percentages. Number of hospitals was assigned to 'x' and case number percentages was assigned to 'y'. Using the linregress function from the scipy.stats library, a linear regression was performed to find the slope, intercept, and R-value between the assigned 'x' and 'y' values. A trend line was also created and the titles of the graph were assigned using pyplot. The R value was also found and displayed using pyplot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

file_path = 'covid_data.csv'  
data = pd.read_csv(file_path)

filtered_data = data[data['county'] != 'Statewide'].copy()

filtered_data['cases_percentage'] = (filtered_data['cases_reported'] / filtered_data['population']) * 100

x = filtered_data['hospitals']
y = filtered_data['cases_percentage']

slope, intercept, r_value, p_value, std_err = linregress(x, y)

trend_line = slope * x + intercept

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7, color='blue', label='Data Points')

plt.plot(x, trend_line, color='red', label=f'Trend Line (R={r_value:.2f})')

plt.title("Case Numbers (%) vs. Hospitals per County", fontsize=14)
plt.xlabel("Number of Hospitals", fontsize=12)
plt.ylabel("Case Numbers (%)", fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Graph & Code Explanation**

For the scatter plot analysis, data was loaded from a CSV file and named data. The 'Statewide' portion of the data was taken out as it was the total amounts. Number of hospitals was assigned to 'x' and fatality rate percentages was assigned to 'y'. Using the linregress function from the scipy.stats library, a linear regression was performed to find the slope, intercept, and R-value between the assigned 'x' and 'y' values. A trend line was also created and the titles of the graph were assigned using pyplot. The R value was also found and displayed using pyplot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

file_path = 'covid_data.csv'  
data = pd.read_csv(file_path)

filtered_data = data[data['county'] != 'Statewide'].copy()

x = filtered_data['hospitals']
y = filtered_data['fatality_rate_pct']

slope, intercept, r_value, p_value, std_err = linregress(x, y)

trend_line = slope * x + intercept

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7, color='blue', label='Data Points')

plt.plot(x, trend_line, color='red', label=f'Trend Line (R={r_value:.2f})')

plt.title("Fatality Rates (%) vs. Hospitals per County", fontsize=14)
plt.xlabel("Number of Hospitals", fontsize=12)
plt.ylabel("Fatality Rate (%)", fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## **Discussion** ##

The study provides valuable insights into the factors influencing COVID-19 case percentages and fatality rates across Washington State counties. However, it is important to acknowledge the limitations of this research and the broader implications of the findings.

**Limitations**

One major limitation is the sheer amount of variables affecting case numbers and fatality rates. For the purpose of this study only a few of the total possible were taken into account. COVID-19's spread and fatality rate is influenced by a vast amount of factors, including individual behaviors, policy interventions, healthcare system capacity, and even culture. While this study focused on a select set of variables such as household size, population demographics, poverty levels, and hospital availability in Washington State counties, these do not capture the full picture.

Another limitation stems from the use of correlation analysis, which identifies relationships between variables, but correlation doesn't mean causation. For example, while vaccination coverage was shown to correlate strongly with lower fatality rates, it doesn't mean it is the sole factor for lower fatality rates. It is important to remember there are relationships between the variables and these vary across counties due to several reasons.

**Implications**

The study also has important implications for public health policy and pandemic preparation, and not just at the county level. The strong correlation between household size and case numbers suggests the need to focus on support for areas with larger households. Also, the significant role of vaccination coverage in reducing fatality rates should be highlighted. More so in the way that it will help with the body dealing with COVID, but doesn't slow down the spread. This is important because there was backlash to the vaccine "not doing anything".

The findings also suggest that socioeconomic factors, such as poverty, should not be overlooked in pandemic responses. Improving access to healthcare and resources can help many in the total population but even more so in disadvantaged communities.

## **Conclusion** ##

The purpose of the study was to find what variables impact COVID-19 case percentages and fatality rates across Washington State counties. The results highlighted that while some factors strongly correlate with case numbers and fatalities, others were surprisingly minor in their impact.

Household size emerged as the most significant variable affecting case numbers, with larger households correlating with higher case amount percentages with an R score of 0.75. However, variables like population density and vaccine coverage had little to no correlation with case percentages, which clashes with preconceptions many have with the COVID pandemic. It was also interesting to note that counties with older populations, 65 years old and above showed lower case numbers. 

Fatality rates were most strongly influenced by vaccination coverage with an R score of -0.60, showing the vaccine did improve the bodies response to COVID-19. Poverty levels had an R score of 0.37 and the percentage of older adults had an R score of 0.45, both positively correlating with fatality rates, showing age and economic status influence fatality rates. Higher population density correlated with lower fatality rates which may be from development and the proximity of resources to individuals.

The role of medical availability, as represented by hospital amounts per county was not as significant as originally thought. While counties with more hospitals showed slightly lower fatality rates and slightly higher case counts the correlation R scores were 0.28 and -0.23 respectively. Showing some significance to the amount of hospitals in every county, but not much.

Overall, these findings emphasize how COVID-19 has been influenced by household size, socioeconomic factors, vaccination, and healthcare access. Showing that there are certain populations that may need greater support through a pandemic.
